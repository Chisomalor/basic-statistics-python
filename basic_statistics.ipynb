BASIC STATISTICS WITH PYTHON

```python
import pandas as pd

# Our agricultural dataset
data = {
    "Crop": ["Maize", "Rice", "Bean", "Tomato", "Yam"],
    "Yield_kg": [1500, 2000, 800, 1200, 3000],
    "Rainfall_mm": [120, 180, 55, 100, 210],
    "Price_Naira": [350, 400, 800, 500, 600]
}

df = pd.DataFrame(data)

# Basic statistics
print("MEAN (Average):")
print(df["Yield_kg"].mean())

print("\nSTANDARD DEVIATION (Spread):")
print(df["Yield_kg"].std())

print("\nMIN and MAX:")
print("Lowest yield:", df["Yield_kg"].min())
print("Highest yield:", df["Yield_kg"].max())

print("\nMEDIAN (Middle Value):")
print(df["Yield_kg"].median())
```

CORRELATION

```python
print("CORRELATION between Rainfall and Yield:")
print(df["Rainfall_mm"].corr(df["Yield_kg"]))
```

VISUALIZING CORRELATION

```python
import matplotlib.pyplot as plt

plt.scatter(df["Rainfall_mm"], df["Yield_kg"], color="green")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Yield (kg)")
plt.title("Rainfall vs Crop Yield Correlation")
plt.show()
```

GROUPBY

```python
# Average yield and price per crop
print("summary statistics per Crop:")
print(df.groupby("Crop")["Yield_kg"].mean())

```
